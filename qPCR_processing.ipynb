```python
import pandas as pd
import os
from numpy import savetxt
import math
```

```python
folder = "C:/Users/Kristine/OneDrive - University of Pittsburgh/Tang Lab/Effect of epigenetics modifiers on HASMCs"
file = "hPGC1A TV2_2 on siTFAM Set1 exp 090821 KS.csv"
result_folder = file.split(".")[0]
if not os.path.exists(result_folder):
    os.mkdir(result_folder)
patient = 4
tp = 2
condition = 6
gene1 = 4
gene2 = 4
gene3 = 4
gene4 = 4
si = True
UR = [5.505205,-5.335725,10.778775]
cq_avg =[16.70006,16.886555,18.310995,17.390445,17.14833,18.19682,20.994525,16.839865,17.123035,16.458015,17.036015,16.24149,17.62357,16.61656,17.37237,18.6399,18.339775,16.57438,17.40458,16.560305,17.145065,17.2988,16.81951,16.508005,16.618895,18.158265,19.03123,17.43201,17.69898,20.543685,18.129595,20.86798,18.78539,18.76647,19.34167,19.40586,19.0859,18.664125,17.647545,17.430615,19.23595,16.173455,16.81611,21.14076,20.770485,19.40658,17.98838,18.886675,]
df = pd.read_csv(os.path.join(folder,file))
```

```python
# Data Cleaning

df = df.replace('Undetermined','NaN')
df['Cq'] = df['Cq'].astype(float)
# df['Well_Initial']= df['Well'].apply(lambda x:x[:1])
```

```python
# Gene1 Calculation

cq1 = df[0:gene1*24]['Cq'].tolist()
df1 = pd.DataFrame()
tmp1 = []
tmp2 = []
for i in range(gene1//2):
    tmp1 = tmp1 + cq1[48*i:48*i+24]
    tmp2 = tmp2 + cq1[48*i+24:48*(i+1)]
df1['Cq1'] = tmp1
df1['Cq2'] = tmp2
df1['Diff'] = df1['Cq1'] - df1['Cq2']
df1['Avg'] = (df1['Cq1'] + df1['Cq2'])/2
df1['Cq avg'] = cq_avg
df1['Delta Cq'] = df1['Avg'] - df1['Cq avg']
RER1 = []
RER2 = []
RER3 = []
for i in range(patient*tp):
    for j in range(condition):
        RER1 = RER1 + [pow(2,df1['Delta Cq'][i*condition] - df1['Delta Cq'][i*condition+j])]
        RER2 = RER2 + [pow(2,UR[0] - df1['Delta Cq'][i*condition+j])]
        if si == True:
            RER3 = RER3 + [pow(2,df1['Delta Cq'][i*condition+2] - df1['Delta Cq'][i*condition+j])]
df1['RER1'] = RER1
ctl1 = df1['RER1'].values.reshape((8,6))
savetxt(os.path.join(result_folder,"ctl_1.csv"), ctl1, delimiter=",")
df1['RER2'] = RER2
URc1 = df1['RER2'].values.reshape((8,6))
savetxt(os.path.join(result_folder,"UR_1.csv"), URc1, delimiter=",")
if si == True:
    df1['RER3'] = RER3
    sic1 = df1['RER3'].values.reshape((8,6))
    savetxt(os.path.join(result_folder,"si_1.csv"), sic1, delimiter=",")
df1.to_csv(os.path.join(result_folder,"gene_1.csv"))
```
